```python
import numpy
import scipy
import math
import matplotlib.pylab
```

```python
x = range(1, 16)
y = list(math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.) for x in x)

# func: math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
matplotlib.pylab.plot(x, y)
matplotlib.pylab.show()
```

```python
A = numpy.array([[1, 1], [1, 15 ** 1]])
b = numpy.array(list(math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.) for x in (1, 15)))
print A, b
```

```python
solution1 = numpy.linalg.solve(A, b)
solution1
```

```python
x = range(1, 16)
y = list(solution1[0] + solution1[1] * x ** 1 for x in x)
matplotlib.pylab.plot(x, y)
matplotlib.pylab.show()
```

```python
A = numpy.array([[1, 1, 1], [1, 8 ** 1, 8 ** 2], [1, 15 ** 1, 15 ** 2]])
b = numpy.array(list(math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.) for x in (1, 8, 15)))
print A, b
```

```python
solution2 = numpy.linalg.solve(A, b)
solution2
```

```python
x = range(1, 16)
y = list(solution2[0] + solution2[1] * x + solution2[2] * x ** 2 for x in x)
matplotlib.pylab.plot(x, y)
matplotlib.pylab.show()
```

```python
A = numpy.array([[1, 1, 1, 1], [1, 4 ** 1, 4 ** 2, 4 ** 3], [1, 10 ** 1, 10 ** 2, 10 ** 3], [1, 15 ** 1, 15 ** 2, 15 ** 3]])
b = numpy.array(list(math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.) for x in (1, 4, 10, 15)))
print A, b
```

```python
solution3 = numpy.linalg.solve(A, b)
solution3
```

```python
x = range(1, 16)
y = list(solution3[0] + solution3[1] * x + solution3[2] * x ** 2 + solution3[3] * x ** 3 for x in x)
matplotlib.pylab.plot(x, y)
matplotlib.pylab.show()
```
