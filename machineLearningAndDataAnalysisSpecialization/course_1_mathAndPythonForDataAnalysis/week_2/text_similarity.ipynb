```python
text = open('sentences.txt')

lines = []
while (True):
    line = text.readline()
    if line == '':
        text.close()
        break
    lines.append(line[0 : -1].lower())

import re

def filterLine(line):
    result = filter(lambda(substr): substr != '', line)
    return result
    
    
lines = list(map(lambda(line): re.split('[^a-z]', line), lines))
lines = list(map(filterLine, lines))
lines
```

```python
wordsDict = []
for line in lines:
    for word in line:
        if (word not in wordsDict):
            wordsDict.append(word)

print len(wordsDict)
print wordsDict
```

```python
import numpy
matrix = numpy.zeros((len(lines), len(wordsDict)))
matrix.shape
```

```python
for idxOfLine, line in enumerate(lines):
    lineVector = numpy.zeros((1, len(wordsDict)))
    for idxOfWord, word in enumerate(wordsDict):
        # print("idxWord: {}, word: {}".format(idxOfWord, word))
        if word in line:
            lineVector[0, idxOfWord] = line.count(word)
    matrix[idxOfLine] = lineVector
    
matrix
        
```

```python
import scipy.spatial

firstNearest = 0
secondNearest = -1

distances = []

for idx, vector in enumerate(matrix):
    distance = scipy.spatial.distance.cosine(matrix[0], vector)
    print distance
    distances.append(distance)

distances = distances[1:]

for idx, dist in enumerate(distances):
    if dist < distances[firstNearest]:
        secondNearest = firstNearest
        firstNearest = idx
        
firstNearest += 1
secondNearest += 1
print firstNearest, secondNearest
```
